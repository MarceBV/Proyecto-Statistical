### Gráficos

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo de gráficos
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Target', data=dataset)
plt.title('Distribución de la Variable Objetivo')
plt.show()
```

### Histograma de las variables numéricas

```python
dataset.hist(bins=30, figsize=(20, 15))
plt.suptitle('Distribución de Variables Numéricas')
plt.show()
```

### Matriz de correlación 

```python
numeric_ds = df.select_dtypes(include=['float64', 'int64'])

corr_matrix = numeric_ds.corr()
plt.figure(figsize=(25, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación')
plt.show()
```

### Boxplot de variables numéricas

```python
plt.figure(figsize=(15, 10))
for i, column in enumerate(df.select_dtypes(include=['float64', 'int64']).columns, 1):
    plt.subplot(5, 4, i)
    sns.boxplot(x='Target', y=column, data=dataset)
    plt.title(f'Boxplot de {column}')
plt.tight_layout()
plt.show()
```

### Pairplot de variables numéricas

```python
sample_columns = df.select_dtypes(include=['float64', 'int64']).columns[:5]  
sns.pairplot(df[sample_columns])
plt.suptitle('Pairplot de Variables Numéricas')
plt.show()
```
